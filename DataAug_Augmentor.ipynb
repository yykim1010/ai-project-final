```python
import Augmentor

#증강 시킬 이미지 폴더 경로
img = Augmentor.Pipeline("/home/piai/Bus/NFC/class")

#좌우 반전
# img.flip_left_right(probability=0.7) 
img.rotate(probability=0.5, max_left_rotation=5, max_right_rotation=5)
#상하 반전
img.flip_top_bottom(probability=0.5)

#모퉁이 왜곡
img.skew_corner(probability=0.5)

#회전&크롭
img.rotate_without_crop(probability=0.5, max_left_rotation=0.8, max_right_rotation=0.8, expand=False, fillcolor=None)

# 왜곡
img.random_distortion(probability=0.5, grid_width=10, grid_height=10, magnitude=10)

#확대
img.zoom_random(probability=0.5, percentage_area=0.8)

#크롭
img.crop_random(probability=0.5, percentage_area=0.2, randomise_percentage_area=False)

#랜덤 지우기
# img.random_erasing(probability=0, rectangle_area=0.3)

#증강 이미지 수
img.sample(100)
```
